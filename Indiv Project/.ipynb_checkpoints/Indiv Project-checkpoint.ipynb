```python
import csv
```

```python
file = open('data.csv', 'rb')
reader = csv.reader(file)
index = 0
Percent_body_fat_from_Siris_1956_equation = []
Age = []
for row in reader:
    if index != 0:
        a = row[0]
        b = a.split()
        Percent_body_fat_from_Siris_1956_equation.append(b[1])
        Age.append(b[2])
        print b      
     
    index += 1
print ""
print "Percent_body_fat_from_Siris_1956_equation: ",
for String in Percent_body_fat_from_Siris_1956_equation:
    print String,
    
print ""

print""
print "Age: ",
for String in Age:
    print String,

file.close()
```

```python
figure = plt.figure()
Chart = figure.add_subplot(111)
Chart.set_title('Percent body fat fromSiris 1956 equation VS Age')
Chart.set_xlabel('Age')
Chart.set_ylabel('Percent body fat from Siris 1956 equation')
Chart.scatter(Age, Percent_body_fat_from_Siris_1956_equation)
figure.savefig("Percent_body_fat_from_Siris_1956_equation_with_titles.png")
```
